```python
import cv2
import numpy as np
import torch
from pathlib import Path

# Load the YOLOv5 model
model = torch.hub.load('ultralytics/yolov5', 'custom', path='yolov5/runs/train/exp/weights/last.pt')

# Define the classes of objects to detect
classes = ['Ball']

# Initialize the Kalman filter
kalman = cv2.KalmanFilter(4, 2)
kalman.measurementMatrix = np.array([[1, 0, 0, 0], [0, 1, 0, 0]], np.float32)
kalman.transitionMatrix = np.array([[1, 0, 1, 0], [0, 1, 0, 1], [0, 0, 1, 0], [0, 0, 0, 1]], np.float32)
kalman.processNoiseCov = np.array([[1e-5, 0, 0, 0], [0, 1e-5, 0, 0], [0, 0, 1e-5, 0], [0, 0, 0, 1e-5]], np.float32)
kalman.measurementNoiseCov = np.array([[1e-1, 0], [0, 1e-1]], np.float32)

# Initialize the video capture
cap = cv2.VideoCapture('/drive/MyDrive/labled-sides.mp4')

# Initialize variables for tracking the ball
state = None
measurement = None
trajectory = []

while True:
    # Read a frame from the video capture
    ret, frame = cap.read()

    if ret:
     # Perform object detection on the frame using YOLOv5
     results = model(frame)
     detections = results.xyxy[0]
    
     # Filter the YOLOv5 output to only include ball detections
     detections = detections[detections[:, 5] == 0]
     detections = detections[detections[:, 4] > 0.5]
    
     if detections.shape[0] > 0:
        # Select the detection with the highest confidence
        detection = detections[detections[:, 4].argmax()]
        x, y, w, h = map(int, detection[0:4])
        center_x = x + w // 2
        center_y = y + h // 2
        
        # Use the Kalman filter to track the ball's position
        if state is None:
            state = np.array([[center_x], [center_y], [0], [0]], np.float32)
            measurement = np.array([[center_x], [center_y]], np.float32)
        else:
            state = kalman.predict()
            measurement = np.array([[center_x], [center_y]], np.float32)
            state = kalman.correct(measurement)
        
        # Add the ball's current position to the trajectory
        trajectory.append((int(state[0]), int(state[1])))
                           
        # Draw the ball's current position on the frame
        cv2.circle(frame, (int(state[0]), int(state[1])), 5, (0, 255, 0), -1)

        # Draw the ball's trajectory on the frame
        if len(trajectory) > 1:
            for i in range(len(trajectory)-1):
                cv2.line(frame, trajectory[i], trajectory[i+1], (0, 255, 0), 2)
         # Display the frame
        cv2.imshow('Frame', frame)
         # Save the frame with the detections
        cv2.imwrite('detections.jpg', frame)

    # Press 'q' to quit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

    else:
     break
    
#Release the video capture and close the window
cap.release()
cv2.destroyAllWindows()



                   
```
